# Problem

Given a string s representing a valid expression, implement a basic calculator to evaluate it, and return the result of the evaluation.

Note: You are not allowed to use any built-in function which evaluates strings as mathematical expressions, such as eval().

#### Example 1:

Input: s = "1 + 1"

Output: 2

#### Example 2:

Input: s = " 2-1 + 2 "

Output: 3

#### Example 3:

Input: s = "(1+(4+5+2)-3)+(6+8)"

Output: 23
 
Constraints:

1 <= s.length <= 3 * $10^5$

s consists of digits, '+', '-', '('  ,  ')', and ' '.

s represents a valid expression.

'+' is not used as a unary operation (i.e., "+1" and "+(2 + 3)" is invalid).

'-' could be used as a unary operation (i.e., "-1" and "-(2 + 3)" is valid).

There will be no two consecutive operators in the input.

Every number and running calculation will fit in a signed 32-bit integer.

```python
class Solution:
    def calculate(self, s: str) -> int:
        stack = []
        current_number = 0 
        current_sign = 1 # +1 for positive, -1 for negative
        result = 0 # outside parenthesis

        for char in s:
            if char.isdigit():
                current_number = current_number * 10 + int(char)
            elif char == '+':
                result += current_sign * current_number
                current_number = 0
                current_sign = 1
            elif char == '-':
                result += current_sign * current_number
                current_number = 0
                current_sign = -1
            elif char == '(':
                # Finish the result and sign before the parenthesis
                stack.append(result)
                stack.append(current_sign)
                # Start a new sub-expression by resetting result and sign
                result = 0
                current_sign = 1
            elif char == ')':
                # Finish the result in the parenthesis
                result += current_sign * current_number
                # Then multiply it with the sign before
                result *= stack.pop()
                # Then add it to the value before the parenthesis
                result += stack.pop()
                # reset the current number
                current_number = 0
        
        return result + current_sign * current_number
```
